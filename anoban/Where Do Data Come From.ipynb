# ***Where Do Data Come From***
------------------

## ***Types of Data***
-----------------

```python
# There are two key types of data
    # Organic / process data
    # Designed data
```

## ***Organic / process data***
-------------------

```python
# These are data, generated by computerized information system / extracted from audio / video recordings.
# These are generated organically by cetrain processes, typically overt time.
# e.g. financial / point of sale (POS) data
# stock market data
# Netflix view history
# web-browser activities
# sports metrics at various levels -> International matches, National matches, College sports, High school sports, middle school sports etc..
# Temperature and pollution sensors
```

```python
# These processes generate massive quantities of data
# Sensationally referred to as BIG DATA.
# Base ball games
# Meals ordered at a particular McDonalds etc..
```

```python
# Given these huge datasets, processing these data requires heavy computational resources.
# Data scientists MINE this data to study relationships and infer summaries.
# However, this mining process itself is very computationally demanding.
```

## ***Designed data collection***
------------------

```python
# These type of data arises from pre-designed data collection
# These are mostly from specific studies specifcally designed to address specific questions
# These could be
# Opinions of a cohort of people on certain topics
# A collection of Tweets, that involve the hashtag #USELECTIONS2022
# These are very rigorously designed data collections opposed to large organic datasets that have accumulated over time.
```

## ***Features of Designed Data***
--------------------

```python
# Sampling from populations, administration of carefully designed questions to select survey samples.
# Datasets are much smaller compared to datasets from organic processes.
# Data is collected for very specific purposes, rather than being simple reflective documentations of an ongoing organic process.
```

## ***Is the data I.I.D?***
---------------

```python
# For analyzing data, regardless of the source, an important question is ->
# Can the data be considered IID?
```

```python
# IID

# I -> Independent
    # Observations (rows / records) in a variable of interest (column) are completely independent of one another.
    # No correlation in different measuements collected from different units in the population.
    # e.g. different baseball game scores, different stock-market exchange rates.
    # etc..
    
# ID -> Identically distributed
    # All the values in a variable arise from a common statistical distribution.
```

```python
ndist = np.random.normal(50, 10, 100000)
plt.hist(ndist, bins = 1000, color = "red", alpha = 0.5)
plt.xlabel("Exam score")
plt.show()
```

```python
# Consider the above bell shaped curve.
# Imagine these represent scores of students in a semester exam.
# Imagine these are independent observations, that each student is providing their independent answers
# i.e no group works, no cheating, no plagiarism etc..
```

```python
# So their scores follow a natural, normal distribution.
# Each student treated independently, that are coming from a common normal distribution.
```
