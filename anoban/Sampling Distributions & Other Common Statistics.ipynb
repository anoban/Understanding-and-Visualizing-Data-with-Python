```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

# ___Sampling Distributions & Other Common Statistics___
----------------

```python
# So far, we have only seen the sampling distribution of means.
# For many common statistics, the sampling distributions are remarkably normal.
```

```python
# Given a large number of samples (many sampling attempts), the sampling distributions of most statistics lean towards a normal
# distribution.
# regardless of the distribution of values in the population.

# This observation is in agreement with central limit theorem.
# This drives design based statistical inferences.
```

## ___Simulation: Pearson correlation coefficient___
-------------

```python
# Pearson correlation coefficient describes the linear association between two continuous variables.
# This is very different from a sample mean.
```

```python
population = np.random.normal(loc = 250, scale = 20, size = 2_000_000)
corrs = {100: [], 500: [], 1000: [], 5000: []}

for ssize in corrs.keys():
    for i in range(1000):
        sample_1 = np.random.choice(population, size = ssize, replace = True)
        sample_2 = np.random.choice(population, size = ssize, replace = True)
        corrs[ssize].append(stats.pearsonr(sample_1, sample_2)[0])
```

```python
fig, ((axes_0, axes_1), (axes_2, axes_3)) = plt.subplots(nrows = 2, ncols = 2)
fig.set_size_inches(10, 10)

axes_0.hist(corrs[100], bins = 20, rwidth = 0.9, color = "red")
axes_1.hist(corrs[500], bins = 20, rwidth = 0.9, color = "red")
axes_2.hist(corrs[1000], bins = 20, rwidth = 0.9, color = "red")
axes_3.hist(corrs[5000], bins = 20, rwidth = 0.9, color = "red")

plt.show()
```

```python
# all of them look bell shapes (normal)
# all of them are nearly centered at 0.00
```
