# ___Sampling Distributions & Other Common Statistics___
----------------

```python
# So far, we have only seen the sampling distribution of means.
# For many common statistics, the sampling distributions are remarkably normal.
```

```python
# Given a large number of samples (many sampling attempts), the sampling distributions of most statistics lean towards a normal
# distribution.
# regardless of the distribution of values in the population.

# This observation is in agreement with central limit theorem.
# This drives design based statistical inferences.
```

## ___Simulation: Pearson correlation coefficient___
-------------

```python
# Pearson correlation coefficient describes the linear association between two continuous variables.
# This is very different from a sample mean.
```

```python
# two populations of random numbers.

population_1 = np.random.randint(low = 250, high = 280, size = 100_000)
population_2 = population_1 + (np.random.random(size = 100_000) * 2)

corrs = {100: [], 500: [], 1000: [], 5000: []}

for ssize in corrs.keys():
    for i in range(1000):
        rands = np.random.randint(low = 0, high = 100_000, size = ssize)
        sample_1 = population_1[rands]
        sample_2 = population_2[rands]
        corrs[ssize].append(stats.pearsonr(sample_1, sample_2)[0])
```

```python
# correlation between the two populations,

stats.pearsonr(population_1, population_2)
```

```python
# there is a strong positive correlation between the two original populations.
```

```python
fig, ((axes_0, axes_1), (axes_2, axes_3)) = plt.subplots(nrows = 2, ncols = 2)
fig.set_size_inches(10, 10)

axes_0.hist(corrs[100], bins = 35, rwidth = 0.9, color = "red")
axes_0.set_title("n = 100")
axes_1.hist(corrs[500], bins = 35, rwidth = 0.9, color = "red")
axes_1.set_title("n = 500")
axes_2.hist(corrs[1000], bins = 35, rwidth = 0.9, color = "red")
axes_2.set_title("n = 1000")
axes_3.hist(corrs[5000], bins = 35, rwidth = 0.9, color = "red")
axes_3.set_title("n = 5000")

plt.show()
```

```python
# all of them look bell shaped (normal)
# As the sample size increases, the distribution becomes more symmetric. 

# The sampling distribution of correlation coefficients are centred around the correlation coefficent of the populations
# THEY'RE CENTRED AROUND 0.9977
```

```python
# These are supposed to give us an unbiased estimate of the whole population.
```
