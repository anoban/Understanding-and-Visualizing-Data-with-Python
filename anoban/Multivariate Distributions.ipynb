```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# ___Multivariate Distributions___
----------------

```python
samples = np.random.multivariate_normal(mean = [4.50, 32.0], cov = [[1.78, 1.78], [1.78, 1.78]], size = 1000)
```

```python
samples.shape
```

```python
fig, (axes_0, axes_1) = plt.subplots(ncols = 2, sharey = True)
fig.set_size_inches(12, 6)

axes_0.hist(samples[:, 0], bins = 15, color = "red", rwidth = 0.9)
axes_0.vlines(x = np.mean(samples[:, 0]), ymin = 0, ymax = 180, color = "black")
axes_0.set_title("X")

axes_1.hist(samples[:, 1], bins = 15, color = "green", rwidth = 0.9)
axes_1.vlines(x = np.mean(samples[:, 1]), ymin = 0, ymax = 180, color = "black")
axes_1.set_title("Y")

plt.show()
```

```python
# These two histograms look approximately normal (and identical)
# Both histograms have identical shapes
# But the values in the vectors are not similar; first one is centered around 4.5 while the second one is centered around  32.
# Since we specified identical standard deviations (cov), they both have the same spread, (hence the shapes)
```
