```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# ___Multivariate Distributions___
----------------

```python
samples = np.random.multivariate_normal(mean = [4.50, 32.0], cov = [[1.78, 1.78], [1.78, 1.78]], size = 1000)
```

```python
samples.shape
```

```python
fig, (axes_0, axes_1) = plt.subplots(ncols = 2, sharey = True)
fig.set_size_inches(12, 6)
ymax = 230

axes_0.hist(samples[:, 0], bins = 15, color = "red", rwidth = 0.9)
axes_0.vlines(x = np.mean(samples[:, 0]), ymin = 0, ymax = ymax, color = "black")
axes_0.set_title("X")

axes_1.hist(samples[:, 1], bins = 15, color = "green", rwidth = 0.9)
axes_1.vlines(x = np.mean(samples[:, 1]), ymin = 0, ymax = ymax, color = "black")
axes_1.set_title("Y")

plt.show()
```

```python
# These two histograms look approximately normal (and identical)
# Both histograms have identical shapes
# But the values in the vectors are not similar; first one is centered around 4.5 while the second one is centered around  32.
# Since we specified identical standard deviations (cov), they both have the same spread, (hence the shapes)
```

```python
fig = plt.figure(figsize = (10, 7.5))
gspec = fig.add_gridspec(nrows = 2, ncols = 2, width_ratios = (4, 1), height_ratios = (1, 4), left = 0.1, right = 0.9,
                         bottom = 0.1, top = 0.9, wspace = 0.05, hspace = 0.05)

axes_0 = fig.add_subplot(gspec[0])
axes_0.hist(samples[:, 0], bins = 15, color = "red", rwidth = 0.9, alpha = 0.7)
axes_0.set_xticks([])

axes_1 = fig.add_subplot(gspec[3])
axes_1.hist(samples[:, 1], bins = 15, color = "green", rwidth = 0.9, orientation = "horizontal", alpha = 0.7)
axes_1.set_yticks([])

axes_2 = fig.add_subplot(gspec[2])
axes_2.plot(samples[:, 0], samples[:, 1], mfc = "orange", mec = "black", marker = "o", linestyle = "none")
axes_2.set_xlabel("$X$")
axes_2.set_ylabel("$Y$")

plt.show()
```

```python
# Here, we can see there's a string linear correlation between the variables X and Y
# Even when the marginal distributions (the histograms) might suggest that there might not be any associations,
# when we plot them together, we can see correlations more clearly.
```
