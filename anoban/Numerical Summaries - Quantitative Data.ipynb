```python
import numpy as np
np.random.seed(2023 - 4 - 9)
import matplotlib.pyplot as plt
import math
import statistics
import pandas as pd
from scipy import stats
```

# ___Numerical Summaries - Quantitative Data___
-------------

```python
heights = np.random.normal(loc = 160, scale = 35, size = 1000)
```

```python
plt.hist(heights, bins = 20, color = "red", rwidth = 0.9)
plt.show()
```

```python
# The above is a roughly bell shaped curve
```

```python
# 5 number summary

# This includes;
    # Minimum
    # Maximum
    # 1st quartile
    # 2nd quartile (median)
    # 3rd quartile
    
# Additionally 
    # Standard deviation
    # Interquartile range (Q3 - Q1)
# may also be included.
```

```python
def five_number_summary(iterable) -> None:
    
    """
    Takes in any Python iterable (must be numerical) and prints out the five number summary of the numeric vector.
    """
    
    assert all([isinstance(x, (int, float)) for x in iterable]), "Non-numeric items found in the iterable!"
    assert len(iterable) > 1, "Input must be a non-empty iterable."
    
    quartiles = statistics.quantiles(iterable)
    
    print(f"Minimum: {min(iterable):.3f}")
    print(f"Maximum: {max(iterable):.3f}")
    print(f"1st quartile: {quartiles[0]:.3f}")
    print(f"2nd quartile: {quartiles[1]:.3f}")
    print(f"3rd quartile: {quartiles[2]:.3f}")
    
    return None
```

```python
five_number_summary(heights)
```

```python
# 1st quartile is often called the 25th percentile.
# 2nd quartile is 50th quantile (the median)
# 3rd quartile is 75th quantile
```

```python
# nth quantile implies n % of all values in the vector falls below the nth quantile value.
# 34% th quantile = 78.25, means 78.25 delimits the vector into a lower 34 % and the upper 66 %.
```

```python
# A cheap trick to get the summary of a numeric vector is;

pd.Series(heights).describe()
```

## ___Interquartile range (IQR)___

```python
# IQR is the difference between the 3rd quartile and the 1st quartile.
# IQR is used as another measure of spread.
```

```python
salaries = stats.skewnorm.rvs(a = 50, loc = 60_000, scale = 10_000, size = 10_000)
rmode = stats.skewnorm.rvs(a = 100, loc = 100_000, scale = 30_000, size = 8_000)
salaries = np.append(salaries, rmode)
```

```python
plt.hist(salaries, bins = 40, color = "orange", rwidth = 0.9)
plt.show()
```

```python
# The above is a bimodal, right skewed distribution.

pd.Series(salaries).describe()
```

```python
five_number_summary(salaries)
```

```python
plt.boxplot(salaries, vert = False)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Mean salary is

print(f"$ {salaries.mean():,.3f}")
```

```python
# Median salary is 

print(f"$ {np.median(salaries):,.3f}")
```

```python
# Generally the impact of outliers on mean is greater than their impact on median.
```

```python
# Note here, that median < mean
# Since there are few extremely high salaries, the mean is shifted right (towards the higher end of x axis)
# However, median is agnostic to the distribution
# Hence, mean > median
```

## ___Exam scores___

```python
scores = stats.skewnorm.rvs(a = -10, loc = 90, scale = 10, size = 1000)

plt.hist(scores, rwidth = 0.9, color = "blue", bins = 30)
plt.show()
```

```python
# Since many very lower scores found (left skewed distribution) at the left,
# We can expect the mean to be lesser than the median
```

```python
print(f"Mean score is {scores.mean():.2f}")
print(f"Median score is {np.median(scores):.2f}")
```

```python
pd.Series(scores).describe()
```
